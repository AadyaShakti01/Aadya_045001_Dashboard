```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact_manual, widgets, Output
from IPython.display import display

# Load data
data = pd.read_csv('superstore_data.csv')  # Replace 'your_dataset.csv' with the actual file name

# Set the color theme to blue
sns.set_palette("Blues_d")
color_palette = sns.color_palette()

# Function to create the customer dashboard with slicers
def create_customer_dashboard(filtered_data):
    fig, axs = plt.subplots(4, 2, figsize=(15, 20), gridspec_kw={'height_ratios': [3, 3, 3, 3], 'width_ratios': [1, 1]})
    # Set the spacing between subplots
    plt.subplots_adjust(hspace=0.5, wspace=0.5)

    # Overall Heading
    fig.suptitle('Customer Dashboard', fontsize=40, fontweight='bold')

    # Graph 1: Age Distribution of Customers
    age_hist = axs[0, 0].hist(filtered_data['Age'], bins=30, color='#3498db', edgecolor='black')
    axs[0, 0].set_title('Age Distribution of Customers', fontsize=15)
    axs[0, 0].set_xlabel('Age')
    axs[0, 0].set_ylabel('Frequency')
    axs[0, 0].set_ylim(0, max(age_hist[0]) + 5)
    axs[0, 0].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 2: Relationship Between Education and Income (Bar Plot)
    bar_plot = sns.barplot(x='Income', y='Education', data=filtered_data, ax=axs[0, 1])
    axs[0, 1].set_title('Average Income Across Education Levels', fontsize=15)
    axs[0, 1].set_xlabel('Average Income')
    axs[0, 1].set_ylabel('Education Level')
    axs[0, 1].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 3: Family Composition (Kidhome and Teenhome)
    family_composition = filtered_data[['Kidhome', 'Teenhome']].sum()
    family_composition.plot(kind='bar', stacked=True, ax=axs[1, 0])
    axs[1, 0].set_title('Family Composition (Kidhome and Teenhome)', fontsize=15)
    axs[1, 0].set_xlabel('Composition')
    axs[1, 0].set_ylabel('Count')
    axs[1, 0].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 4: Customer Recency and Web Visits
    scatter_plot = sns.scatterplot(x='Recency', y='NumWebVisitsMonth', data=filtered_data, ax=axs[1, 1])
    axs[1, 1].set_title('Customer Recency and Web Visits', fontsize=15)
    axs[1, 1].set_xlabel('Recency')
    axs[1, 1].set_ylabel('Number of Web Visits per Month')
    axs[1, 1].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 5: Pie chart of Marital Status distribution
    marital_status_distribution = filtered_data['Marital_Status'].value_counts()
    axs[2, 0].pie(marital_status_distribution, labels=marital_status_distribution.index, autopct='%1.1f%%', colors=sns.color_palette())
    axs[2, 0].set_title('Distribution of Marital Status', fontsize=15)
    axs[2, 0].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 6: Monthly Spending on Products Over Time
    product_categories = filtered_data.columns[9:-6]
    filtered_data['Dt_Customer'] = pd.to_datetime(filtered_data['Dt_Customer'])
    filtered_data['Month'] = filtered_data['Dt_Customer'].dt.month
    monthly_spending = filtered_data.groupby('Month')[product_categories].sum()
    for i, col in enumerate(monthly_spending.columns):
        color = color_palette[i % len(color_palette)]  # Cycle through the color palette
        axs[2, 1].plot(monthly_spending.index, monthly_spending[col], marker='o', linestyle='-', markersize=8, label=col, color=color)

    axs[2, 1].set_title('Monthly Spending on Products Over Time', fontsize=15)
    axs[2, 1].set_xlabel('Month')
    axs[2, 1].set_ylabel('Total Spending')
    axs[2, 1].legend()
    axs[2, 1].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 7: Distribution of Responses to Marketing Campaigns
    response_distribution = filtered_data['Response'].value_counts()
    axs[3, 0].pie(response_distribution, labels=response_distribution.index, autopct='%1.1f%%', colors=sns.color_palette())
    axs[3, 0].set_title('Distribution of Responses to Marketing Campaigns', fontsize=15)
    axs[3, 0].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Graph 8: Customer Complaints
    complaints_distribution = filtered_data['Complain'].value_counts()
    axs[3, 1].pie(complaints_distribution, labels=complaints_distribution.index, autopct='%1.1f%%', colors=sns.color_palette())
    axs[3, 1].set_title('Distribution of Customer Complaints', fontsize=15)
    axs[3, 1].annotate(f'Count: {len(filtered_data)}', xy=(0.5, 0.9), xycoords="axes fraction", ha="center", va="center", bbox=dict(boxstyle="round", alpha=0.1))

    # Overall Outline
    for ax in axs.flatten():
        ax.spines['top'].set_visible(True)
        ax.spines['right'].set_visible(True)

    # Draw boundary around all charts
    fig.patch.set_facecolor('#d9d9d9')  # Background color of the entire figure
    fig.patch.set_edgecolor('black')   # Color of the boundary
    fig.patch.set_linewidth(7)         # Width of the boundary
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the layout to prevent cutoff

# Define education levels and marital statuses
education_levels = ['', 'Graduation', 'PhD', 'Master', 'Basic']
marital_statuses = ['', 'Single', 'Married', 'Divorced', 'Together']

# Create interactive dropdowns
education_dropdown = widgets.Dropdown(options=education_levels, value='', description='Education Level')
marital_dropdown = widgets.Dropdown(options=marital_statuses, value='', description='Marital Status')

# Use interact_manual to create dynamic dropdowns
output = Output()
display(output)

def update_dashboard(education_level, marital_status):
    with output:
        plt.close()  # Close previous plots
        filtered_data = data.copy()
        if education_level:
            filtered_data = filtered_data[filtered_data['Education'] == education_level]
        if marital_status:
            filtered_data = filtered_data[filtered_data['Marital_Status'] == marital_status]

        create_customer_dashboard(filtered_data)
        plt.show()

interact_manual(update_dashboard, education_level=education_dropdown, marital_status=marital_dropdown);

#code taken from Chatgpt after multiple iterations
```
